```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Crear DataFrame con los datos del ejercicio
datos = {
"Cliente": [1, 2, 3, 4, 5],
"Edad": [25, 40, 32, 22, 35],
"Plan": ["Básico", "Premium", "Estándar", "Básico", "Premium"],
"Consumo_GB": [5, 50, 10, 7, 45],
"Reclamos": [1, 0, 2, 3, 0],
"Estado_Cuenta": ["Pagado", "Pagado", "Moroso", "Moroso", "Pagado"]
}
df = pd.DataFrame(datos)
```

```python
# Inspeccionar los datos
print(df.info())
print(df.describe())
```

```python
# Ver distribución de consumo de datos
sns.histplot(df["Consumo_GB"], bins=5, kde=True)
plt.show()
```

```python
# Gráfico de barras de Estado de Cuenta
sns.countplot(x=df["Estado_Cuenta"])
plt.show()
```

Ejecuté el código base provisto en el material dentro de un entorno Jupyter Notebook. A partir de la inspección inicial del DataFrame, verifiqué que los datos estaban completos y sin valores nulos. Luego realicé un análisis estadístico descriptivo y visualicé la distribución del consumo de datos mediante el histograma, así como el estado de cuenta con el gráfico de barras. En el histograma se observa una diferencia clara entre dos grupos de clientes: uno con consumo bajo (entre 5 y 10 GB) y otro con consumo mucho más alto (45 y 50 GB). En el gráfico de barras se ve que tres clientes tienen la cuenta al día, mientras que dos figuran como morosos. Al observar los reclamos (en el dataframe), se observa que justamente los clientes morosos son quienes más reclamos tienen y consumen menos, lo cual podría indicar la existencia de una relación entre los planes de menor capacidad de datos y un mayor nivel de insatisfacción por parte de quienes los utilizan.
